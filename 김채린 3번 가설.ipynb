```python
# 0) seaborn 경고 해결: ci 대신 errorbar=None 쓰세요 (아래 예제에도 반영됨)

# 1) 한글 폰트 설치 (Noto + Nanum)
!apt-get -qq -y install fonts-noto-cjk fonts-nanum > /dev/null

# 2) 폰트 등록(재시작 없이 바로 적용): 설치된 경로에서 폰트 찾아 addfont
import os, glob
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

candidates = []
candidates += glob.glob('/usr/share/fonts/**/Noto*Sans*CJK*.*', recursive=True)
candidates += glob.glob('/usr/share/fonts/**/NotoSansKR*.*', recursive=True)
candidates += glob.glob('/usr/share/fonts/**/NanumGothic*.*', recursive=True)

for p in candidates:
    try:
        fm.fontManager.addfont(p)
    except Exception:
        pass

# 3) 실제 등록된 폰트 이름 중에서 선호 순서대로 선택
available = sorted(set(f.name for f in fm.fontManager.ttflist))
preferred = None
for name in [
    "Noto Sans CJK KR",  # 최선
    "Noto Sans KR",      # 대안
    "NanumGothic",       # 대안
    "Nanum Gothic",      # 대안 (일부 시스템 표기)
]:
    if name in available:
        preferred = name
        break

# 4) 폰트 적용
if preferred is None:
    # 그래도 못 찾으면 아무거나 Noto/Nanum 계열 중 첫 번째 선택
    for n in available:
        if ("Noto" in n) or ("Nanum" in n):
            preferred = n
            break

plt.rcParams['font.family'] = preferred if preferred else 'DejaVu Sans'
mpl.rcParams['axes.unicode_minus'] = False

print("사용 폰트:", plt.rcParams['font.family'])

# 5) 한글 테스트 (네모 없이 보여야 정상)
plt.figure()
plt.title("한글 폰트 테스트 - 카테고리별 평균 리뷰 점수")
plt.xlabel("상품카테고리")
plt.ylabel("평균 리뷰 점수")
plt.plot([1,2,3], [4.2,4.3,4.1])
plt.show()
```

- H3-3: 소비재(패션, 뷰티, 식품)는 기대치가 상대적으로 낮아 리뷰 점수가 높게 나올 가능성이 있다 --> 부합한다. 패션/의류/악세서리, 뷰티/헬스, 식품/음료 전반적으로 높은 점수

-즉, 카테고리별로 극적인 차이는 없지만, 내구재는 비교적 낮고 소비재는 비교적 높은 패턴이 나타난다.

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
```

```python
X = df[["상품카테고리", "상품가격합계", "배송비합계", "총결제금액"]]
y = df["리뷰점수"]
```

```python
X = pd.get_dummies(X, columns=["상품카테고리"], drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
model = RandomForestClassifier(n_estimators=200, random_state=42, class_weight="balanced")
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=sorted(y.unique()), yticklabels=sorted(y.unique()))
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
importances = pd.Series(model.feature_importances_, index=X.columns)
importances.sort_values(ascending=False).head(10).plot(kind="bar", figsize=(8,4))
plt.title("Feature Importance (상위 10개)")
plt.show()
```

모델이 불만족 리뷰를 거의 다 놓치고 있음.

```python
df["리뷰긍정"] = df["리뷰점수"].apply(lambda x: 1 if x >= 4 else 0)

X = df[["상품카테고리","상품가격합계","배송비합계","총결제금액"]]
X = pd.get_dummies(X, columns=["상품카테고리"], drop_first=True)

y = df["리뷰긍정"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
model = RandomForestClassifier(random_state=42, n_estimators=200, class_weight="balanced")
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
#부정리뷰는 놓치고 긍정리뷰만 잘 맞춤
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["부정","긍정"], yticklabels=["부정","긍정"])
plt.xlabel("예측값")
plt.ylabel("실제값")
plt.title("리뷰 긍정/부정 예측 혼동행렬")
plt.show()
#부정/부정 수가 적음
```

```python
importances = pd.Series(model.feature_importances_, index=X.columns).sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=importances, y=importances.index)
plt.title("변수 중요도")
plt.show()
```
