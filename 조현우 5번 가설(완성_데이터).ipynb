```python
df['구매일시'] = pd.to_datetime(df['구매일시'])
df['실제배송일'] = pd.to_datetime(df['실제배송일'])
df['예상배송일'] = pd.to_datetime(df['예상배송일'])

# datetiem 컬럼 전처리
df['예상_소요일수'] = (df['예상배송일'] - df['구매일시']).dt.days
df['실제_소요일수'] = (df['실제배송일'] - df['구매일시']).dt.days
df['예상대비_지연일수'] = (df['실제배송일'] - df['예상배송일']).dt.days
df['예상대비_지연여부'] = (df['예상대비_지연일수'] > 0).astype(int)

# 리뷰 작성일 처리 (상품을 받은 뒤 리뷰 작성까지 걸린 시간)
df['리뷰작성일'] = pd.to_datetime(df['리뷰작성일'])
df['리뷰_작성지연일수'] = (df['리뷰작성일'] - df['실제배송일']).dt.days

# 리뷰 내용 여부 컬럼 / 리뷰_길이 컬럼 생성
df['리뷰내용여부'] = df['리뷰내용'].notna().astype(int)
df["리뷰_길이"] = df["리뷰내용"].fillna("").apply(len)

# 나머지 삭제
df = df.drop(columns = ['예상배송일', '실제배송일', '리뷰작성일', '배송지연일수', '배송지연여부', '리뷰내용'])


pd.set_option('display.max_columns', None)
df.head()
```

```python
cols = df.select_dtypes(include = 'O').columns
cols

df = pd.get_dummies(df, columns=cols, drop_first=False)
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import accuracy_score, mean_absolute_error
from sklearn.metrics import f1_score
```

```python
X_tr, X_val, y_tr, y_val = train_test_split(df.drop('리뷰점수', axis = 1),
                                            df['리뷰점수'],
                                            test_size = 0.2,
                                            random_state = 2025)
```

```python
X_tr.shape, X_val.shape, y_tr.shape, y_val.shape
```

```python
model = LinearRegression()
model.fit(X_tr, y_tr)

y_pred = model.predict(X_val)
```

```python
print("MSE:", mean_squared_error(y_val, y_pred))
print("R2:", r2_score(y_val, y_pred))
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, mean_absolute_error, f1_score
from collections import Counter

major = Counter(y_tr).most_common(1)[0][0]
y_pred_major = np.full_like(y_val, fill_value=major)

print("=== Majority Baseline ===")
print("Major class:", major)
print("Accuracy:", accuracy_score(y_val, y_pred_major))
print("MAE:", mean_absolute_error(y_val, y_pred_major))
print("F1 (weighted):", f1_score(y_val, y_pred_major, average="weighted"))
print("Accuracy@±1:", (np.abs(y_val - y_pred_major) <= 1).mean())
```

```python
# 리뷰 1~3 낮음 / 4~5 높음 / 이진분류

import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix

y_tr_bin = (y_tr >= 4).astype(int) # 4~5 높음
y_val_bin = (y_val >= 4).astype(int) # 1~3 낮음

logreg = LogisticRegression(max_iter=1000, random_state=2025)
logreg.fit(X_tr, y_tr_bin)
y_pred_logreg = logreg.predict(X_val)

print('--Logistic--')
print("Accuracy:", accuracy_score(y_val_bin, y_pred_logreg))
print("F1:", f1_score(y_val_bin, y_pred_logreg))
print("Confusion Matrix:\n", confusion_matrix(y_val_bin, y_pred_logreg))

rf = RandomForestClassifier(random_state=2025)
rf.fit(X_tr, y_tr_bin)
y_pred_rf = rf.predict(X_val)

print('--RandomForest--')
print("Accuracy:", accuracy_score(y_val_bin, y_pred_rf))
print("F1:", f1_score(y_val_bin, y_pred_rf))
print("Confusion Matrix:\n", confusion_matrix(y_val_bin, y_pred_rf))
```

```python
 # 리뷰 1~3 낮음 / 4~5 높음 / 이진분류 / 낮은 리뷰에 가중치 추가


y_tr_bin = (y_tr >= 4).astype(int) # 4~5 높음
y_val_bin = (y_val >= 4).astype(int) # 1~3 낮음


## ***최종 모형 선택*** ##
logreg = LogisticRegression(class_weight='balanced',max_iter=1000, random_state=2025)
logreg.fit(X_tr, y_tr_bin)
y_pred_logreg = logreg.predict(X_val)

print('--Logistic--')
print("Accuracy:", accuracy_score(y_val_bin, y_pred_logreg))
print("F1:", f1_score(y_val_bin, y_pred_logreg))
print("Confusion Matrix:\n", confusion_matrix(y_val_bin, y_pred_logreg))

rf = RandomForestClassifier(class_weight='balanced',random_state=2025)
rf.fit(X_tr, y_tr_bin)
y_pred_rf = rf.predict(X_val)

print('--RandomForest--')
print("Accuracy:", accuracy_score(y_val_bin, y_pred_rf))
print("F1:", f1_score(y_val_bin, y_pred_rf))
print("Confusion Matrix:\n", confusion_matrix(y_val_bin, y_pred_rf))
```

```python
# 리뷰 1~3 낮음 / 4~5 높음 / 이진분류 / 오버 샘플링으로 데이터 분포 맞춤

y_tr_bin  = (y_tr  >= 4).astype(int)
y_val_bin = (y_val >= 4).astype(int)

from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42)
X_res, y_res_bin = sm.fit_resample(X_tr, y_tr_bin)

logreg = LogisticRegression(random_state=2025)
logreg.fit(X_res, y_res_bin)
y_pred_logreg = logreg.predict(X_val)

print('--Logistic--')
print("Accuracy:", accuracy_score(y_val_bin, y_pred_logreg))
print("F1:", f1_score(y_val_bin, y_pred_logreg))
print("Confusion Matrix:\n", confusion_matrix(y_val_bin, y_pred_logreg))

# RandomForest
rf = RandomForestClassifier(random_state=2025)
rf.fit(X_res, y_res_bin)
y_pred_rf = rf.predict(X_val)

rf.fit(X_res, y_res_bin)
y_pred_rf = rf.predict(X_val)

print('--RandomForest__')
print("Accuracy:", accuracy_score(y_val_bin, y_pred_rf))
print("F1:", f1_score(y_val_bin, y_pred_rf))
print("Confusion Matrix:\n", confusion_matrix(y_val_bin, y_pred_rf))
```
